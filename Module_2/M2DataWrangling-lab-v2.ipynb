### 1. Load the Dataset


<h5>1.1 Import necessary libraries and load the dataset.</h5>


Ensure the dataset is loaded correctly by displaying the first few rows.


```python
# Import necessary libraries
import pandas as pd

# Load the Stack Overflow survey data
dataset_url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv"
df = pd.read_csv(dataset_url)

# Display the first few rows
print(df.head())
```

#### 2. Explore the Dataset


<h5>2.1 Summarize the dataset by displaying the column data types, counts, and missing values.</h5>


```python
import numpy as np
# Write your code here
print(df.dtypes)
for column in df.columns:
    print(df[column].value_counts(), "/n")
df.replace("?", np.nan, inplace=True)
missing_data = df.isnull()
for column in missing_data.columns.values.tolist():
    print(missing_data[column].value_counts(), "/n")
```

<h5>2.2 Generate basic statistics for numerical columns.</h5>


```python
# Write your code here
df.describe()
```

### 3. Identifying and Removing Inconsistencies


<h5>3.1 Identify inconsistent or irrelevant entries in specific columns (e.g., Country).</h5>


```python
# Write your code here
pd.set_option('display.max_rows', None)
df[['Country']].value_counts()
# All are consistent and relevant
# If an irrevant entry was to exist, it would be removed
# If an inconsistent entry was to exist, it would to converted to a consistent one by changing its name
```

<h5>3.2 Standardize entries in columns like Country or EdLevel by mapping inconsistent values to a consistent format.</h5>


```python
## Write your code here
print(df[['EdLevel']].value_counts())
# I will convert "Secondary school (e.g. American high school, German Realschule or Gymnasium, etc.)"
# to "Secondary school (e.g. High school) for worldwide consistency"
df['EdLevel'] = df['EdLevel'].replace("Secondary school (e.g. American high school, German Realschule or Gymnasium, etc.)", "Secondard school (e.g. High School)")
print(df[['EdLevel']].value_counts())
```

### 4. Encoding Categorical Variables


<h5>4.1 Encode the Employment column using one-hot encoding.</h5>


```python
## Write your code here
dummy_variable = pd.get_dummies(df["Employment"])
df = pd.concat([df, dummy_variable], axis=1)
```

### 5. Handling Missing Values


<h5>5.1 Identify columns with the highest number of missing values.</h5>


```python
## Write your code here
for column in missing_data.columns.values.tolist():
    print(missing_data[column].value_counts(), "/n")
# TechDoc, BuildvsBuy, TechEndorse, CompTotal and Database Admired are examples of columns with high missing values
```

<h5>5.2 Impute missing values in numerical columns (e.g., `ConvertedCompYearly`) with the mean or median.</h5>


```python
## Write your code here
avg_comp = df['ConvertedCompYearly'].mean()
df['ConvertedCompYearly']= df['ConvertedCompYearly'].replace(np.nan, avg_comp)
missingConvertedCompYearly = df['ConvertedCompYearly'].isnull()
print(missingConvertedCompYearly.value_counts())
```

<h5>5.3 Impute missing values in categorical columns (e.g., `RemoteWork`) with the most frequent value.</h5>


```python
## Write your code here
common_value = df['RemoteWork'].value_counts().idxmax()
df['RemoteWork'] = df['RemoteWork'].replace(np.nan, common_value)
missingRemoteWork = df['RemoteWork'].isnull()
print(missingRemoteWork.value_counts())
```

### 6. Feature Scaling and Transformation


<h5>6.1 Apply Min-Max Scaling to normalize the `ConvertedCompYearly` column.</h5>


```python
## Write your code here
df['ConvertedCompYearlyNormalised'] = (df['ConvertedCompYearly'] - df['ConvertedCompYearly'].min()) / (df['ConvertedCompYearly'].max() - df['ConvertedCompYearly'].min())
df[['ConvertedCompYearly','ConvertedCompYearlyNormalised']].head(10)
```

<h5>6.2 Log-transform the ConvertedCompYearly column to reduce skewness.</h5>


```python
## Write your code here
df['ConvertedCompYearlyLog'] = np.log(df['ConvertedCompYearly'] + 1)
df[['ConvertedCompYearly','ConvertedCompYearlyLog']].head(10)
```

### 7. Feature Engineering


<h5>7.1 Create a new column `ExperienceLevel` based on the `YearsCodePro` column:</h5>


```python
## Write your code here
print(df['YearsCodePro'].value_counts())

def experience(y):
    if y == "?":
        return np.nan
    elif y == "Less than 1 year":
        return "Beginner"
    elif int(y) >= 2 and int(y) <= 5:
        return "Intermediate"
    elif int(y) >= 6 and int(y) <= 15:
        return "Advanced"
    else:
        return "Expert"

df['YearsCodePro'].replace("More than 50 years", 50, inplace=True)
df['ExperienceLevel'] = df['YearsCodePro'].apply(experience)
print(df['ExperienceLevel'].value_counts())
```
