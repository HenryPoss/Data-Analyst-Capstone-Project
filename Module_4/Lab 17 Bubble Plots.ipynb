### Task 1: Exploring Data Distributions Using Bubble Plots


#### 1. Bubble Plot for Age vs. Frequency of Participation


- Visualize the relationship between respondents’ age and their participation frequency (`SOPartFreq`) using a bubble plot.

- Use the size of the bubbles to represent their job satisfaction (`JobSat`).


```python
##Write your code here
dfAgePart = df[['Age', 'SOPartFreq', 'JobSat']]
dfAgePart.dropna(axis=0, inplace=True)

dfAgePart = dfAgePart[dfAgePart['Age'] != 'Prefer not to say']
dfAgePart.replace('Under 18 years old', '18 years old', inplace=True)
dfAgePart.sort_values(by='Age', inplace=True)

norm_sat = (dfAgePart['JobSat'] - dfAgePart['JobSat'].min()) / (dfAgePart['JobSat'].max() - dfAgePart['JobSat'].min())

dfAgePart.plot(kind='scatter', x='Age', y='SOPartFreq', alpha=0.5, s=norm_sat * 300 + 10)
plt.title('Age vs participation sized by job satisfaction')
plt.xticks(rotation=90)
plt.show()
```

#### 2. Bubble Plot for Compensation vs. Job Satisfaction


-Visualize the relationship between yearly compensation (`ConvertedCompYearly`) and job satisfaction (`JobSat`).

- Use the size of the bubbles to represent respondents’ age.


```python
##Write your code here
dfAgeSat = df[['ConvertedCompYearly', 'JobSat', 'Age']]
dfAgeSat.dropna(axis=0, inplace=True)

dfAgeSat.replace('Under 18 years old', 18, inplace=True)
dfAgeSat.replace('18-24 years old', 21, inplace=True)
dfAgeSat.replace('25-34 years old', 29.5, inplace=True)
dfAgeSat.replace('35-44 years old', 39.5, inplace=True)
dfAgeSat.replace('45-54 years old', 49.5, inplace=True)
dfAgeSat.replace('55-64 years old', 59.5, inplace=True)
dfAgeSat.replace('65 years or older', 65, inplace=True)

dfAgeSat = dfAgeSat[dfAgeSat != 'Prefer not to say']
dfAgeSat = dfAgeSat.astype(float)

norm_age = (dfAgeSat['Age'] - dfAgeSat['Age'].min()) / (dfAgeSat['Age'].max() - dfAgeSat['Age'].min())

dfAgeSat.plot(kind='scatter', x='ConvertedCompYearly', y='JobSat', alpha=0.5, s=norm_age * 500 + 10)
plt.title('Compensation vs job satisfaction weighted by age')
plt.show()
```

### Task 2: Analyzing Relationships Using Bubble Plots


#### 1. Bubble Plot of Technology Preferences by Age

- Visualize the popularity of programming languages respondents have worked with (`LanguageHaveWorkedWith`) across age groups.

- Use bubble size to represent the frequency of each language.



```python
##Write your code here
dfLanguageAge = df[['Age', 'LanguageHaveWorkedWith']]
dfLanguageAge.dropna(axis=0, inplace=True)

dfLanguageAge['LanguageHaveWorkedWith'] = dfLanguageAge['LanguageHaveWorkedWith'].str.split(';')

dfLanguageAge = dfLanguageAge.explode('LanguageHaveWorkedWith')

top10Languages = dfLanguageAge['LanguageHaveWorkedWith'].value_counts().head(10).index.tolist()
dfLanguageAge = dfLanguageAge[dfLanguageAge['LanguageHaveWorkedWith'].isin(top10Languages)]

dfLanguageAge = dfLanguageAge[dfLanguageAge['Age'] != 'Prefer not to say']
dfLanguageAge.replace('Under 18 years old', '18 years old', inplace=True)
dfLanguageAge.sort_values(by='Age', inplace=True)

dfFrequency = dfLanguageAge.groupby(['Age', 'LanguageHaveWorkedWith']).size().reset_index(name='Total')

weight = dfFrequency['Total']

dfFrequency.plot(kind='scatter', x='Age', y='LanguageHaveWorkedWith', alpha=0.5, s=weight * 0.2)
plt.title('Languages used vs age weighed by frequency')
plt.xticks(rotation=90)
plt.show()
```

#### 2. Bubble Plot for Preferred Databases vs. Job Satisfaction

- Explore the relationship between preferred databases (`DatabaseWantToWorkWith`) and job satisfaction.

- Use bubble size to indicate the number of respondents for each database.


```python
##Write your code here
dfDatabaseSat = df[['DatabaseWantToWorkWith', 'JobSat']]
dfDatabaseSat.dropna(axis=0, inplace=True)

dfDatabaseSat['DatabaseWantToWorkWith'] = dfDatabaseSat['DatabaseWantToWorkWith'].str.split(';')

dfDatabaseSat = dfDatabaseSat.explode('DatabaseWantToWorkWith')

top10Database = dfDatabaseSat['DatabaseWantToWorkWith'].value_counts().head(10).index.tolist()
dfDatabaseSat = dfDatabaseSat[dfDatabaseSat['DatabaseWantToWorkWith'].isin(top10Database)]

dfFrequency2 = dfDatabaseSat.groupby(['DatabaseWantToWorkWith', 'JobSat']).size().reset_index(name='Total')

weight2 = dfFrequency2['Total']

dfFrequency2.plot(kind='scatter', x='DatabaseWantToWorkWith', y='JobSat', alpha=0.5, s=weight2 * 0.5)
plt.title('Database want to work with vs job satisfaction')
plt.xticks(rotation=90)
plt.show()
```

### Task 3: Comparing Data Using Bubble Plots


#### 1. Bubble Plot for Compensation Across Developer Roles

- Visualize compensation (`ConvertedCompYearly`) across different developer roles (`DevType`).

- Use bubble size to represent job satisfaction.


```python
##Write your code here
dfDevComp = df[['DevType', 'ConvertedCompYearly', 'JobSat']]
dfDevComp.dropna(axis=0, inplace=True)

top10Types = dfDevComp['DevType'].value_counts().head(10).index.tolist()
dfDevComp = dfDevComp[dfDevComp['DevType'].isin(top10Types)]

norm_sat = (dfDevComp['JobSat'] - dfDevComp['JobSat'].min()) / (dfDevComp['JobSat'].max() - dfDevComp['JobSat'].min())

dfDevComp.plot(kind='scatter', x='DevType', y='ConvertedCompYearly', alpha=0.5, s=norm_sat * 200)
plt.title('Compensation based on developer type weighed by job satisfaction')
plt.xticks(rotation=90)
plt.show()
```

#### 2. Bubble Plot for Collaboration Tools by Age

- Visualize the relationship between the collaboration tools used (`NEWCollabToolsHaveWorkedWith`) and age groups.

- Use bubble size to represent the frequency of tool usage.


```python
##Write your code here
dfCollaborationAge = df[['Age', 'NEWCollabToolsHaveWorkedWith']]
dfCollaborationAge.dropna(axis=0, inplace=True)

dfCollaborationAge['NEWCollabToolsHaveWorkedWith'] = dfCollaborationAge['NEWCollabToolsHaveWorkedWith'].str.split(';')

dfCollaborationAge = dfCollaborationAge.explode('NEWCollabToolsHaveWorkedWith')

top10Collaboration = dfCollaborationAge['NEWCollabToolsHaveWorkedWith'].value_counts().head(10).index.tolist()
dfCollaborationAge = dfCollaborationAge[dfCollaborationAge['NEWCollabToolsHaveWorkedWith'].isin(top10Collaboration)]

dfCollaborationAge = dfCollaborationAge[dfCollaborationAge['Age'] != 'Prefer not to say']
dfCollaborationAge.replace('Under 18 years old', '18 years old', inplace=True)
dfCollaborationAge.sort_values(by='Age', inplace=True)

dfCollaborationAge.dropna(axis=0, inplace=True)

dfFrequency3 = dfCollaborationAge.groupby(['Age', 'NEWCollabToolsHaveWorkedWith']).size().reset_index(name='Total')

weight3 = dfFrequency3['Total']

dfFrequency3.plot(kind='scatter', x='Age', y='NEWCollabToolsHaveWorkedWith', alpha=0.5, s=weight3 * 0.2)
plt.title('Collaboration tools used vs age weighed by frequency')
plt.xticks(rotation=90)
plt.show()
```

### Task 4: Visualizing Technology Trends Using Bubble Plots


#### 1. Bubble Plot for Preferred Web Frameworks vs. Job Satisfaction

- Explore the relationship between preferred web frameworks (`WebframeWantToWorkWith`) and job satisfaction.

- Use bubble size to represent the number of respondents.



```python
##Write your code here
dfFrameworkSat = df[['WebframeWantToWorkWith', 'JobSat']]
dfFrameworkSat.dropna(axis=0, inplace=True)

dfFrameworkSat['WebframeWantToWorkWith'] = dfFrameworkSat['WebframeWantToWorkWith'].str.split(';')

dfFrameworkSat = dfFrameworkSat.explode('WebframeWantToWorkWith')

top10Framework = dfFrameworkSat['WebframeWantToWorkWith'].value_counts().head(10).index.tolist()
dfDatabaseSat = dfFrameworkSat[dfFrameworkSat['WebframeWantToWorkWith'].isin(top10Framework)]

dfFrequency3 = dfDatabaseSat.groupby(['WebframeWantToWorkWith', 'JobSat']).size().reset_index(name='Total')

weight3 = dfFrequency3['Total']

dfFrequency3.plot(kind='scatter', x='WebframeWantToWorkWith', y='JobSat', alpha=0.5, s=weight3 * 0.5)
plt.title('Web framework want to work with vs job satisfaction')
plt.xticks(rotation=90)
plt.show()
```

#### 2. Bubble Plot for Admired Technologies Across Countries

- Visualize the distribution of admired technologies (`LanguageAdmired`) across different countries (`Country`).

- Use bubble size to represent the frequency of admiration.



```python
##Write your code here
dfLanguageCountry = df[['LanguageAdmired', 'Country']]
dfLanguageCountry.dropna(axis=0, inplace=True)

dfLanguageCountry['LanguageAdmired'] = dfLanguageCountry['LanguageAdmired'].str.split(';')

dfLanguageCountry = dfLanguageCountry.explode('LanguageAdmired')

top10Languages = dfLanguageCountry['LanguageAdmired'].value_counts().head(10).index.tolist()
dfLanguageCountry = dfLanguageCountry[dfLanguageCountry['LanguageAdmired'].isin(top10Languages)]

top10Countries = dfLanguageCountry['Country'].value_counts().head(10).index.tolist()
dfLanguageCountry = dfLanguageCountry[dfLanguageCountry['Country'].isin(top10Countries)]

dfFrequency4 = dfLanguageCountry.groupby(['LanguageAdmired', 'Country']).size().reset_index(name='Total')

weight4 = dfFrequency4['Total']

dfFrequency4.plot(kind='scatter', x='LanguageAdmired', y='Country', alpha=0.5, s=weight4 * 0.5)
plt.title('Language Admired vs Country')
plt.xticks(rotation=90)
plt.show()
```
