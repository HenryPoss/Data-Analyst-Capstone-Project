### Task 1: Visualizing Data Composition with Pie Charts


##### 1.1 Create a Pie Chart of the Top 5 Databases Respondents Want to Work With


In the survey data, the `DatabaseWantToWorkWith` column lists the databases that respondents wish to work with. Let’s visualize the top 5 most-desired databases in a pie chart.



```python
##Write your code here
top5Databasedf = df['DatabaseWantToWorkWith'].reset_index(name='Database')
top5Databasedf.dropna(axis=0, inplace=True)

top5Databasedf['Database'] = top5Databasedf['Database'].str.split(';')

top5Databasedf = top5Databasedf.explode('Database')

top5Databasedf = top5Databasedf.groupby('Database').count().sort_values(by='index', ascending=False).head(5)

top5Databasedf = top5Databasedf.rename(columns={'index' : 'Database'})
top5Databasedf['Database'].plot(kind='pie', figsize=(12, 8), autopct='%1.1f%%', startangle=90)
plt.title('Top 5 databases respondants want to work with')
plt.axis('equal')

plt.legend(labels=top5Databasedf.index, loc='upper right')

plt.show()
```

The `DevType` column lists the developer types for respondents. We’ll examine the distribution by showing the top 5 developer roles in a pie chart.



```python
##Write your code here
top5Typesdf = df['DevType'].reset_index(name='Type')
top5Typesdf.dropna(axis=0, inplace=True)

top5Typesdf = top5Typesdf.groupby('Type').count().sort_values(by='index', ascending=False).head(5)

top5Typesdf = top5Typesdf.rename(columns={'index' : 'Type'})
top5Typesdf['Type'].plot(kind='pie', figsize=(12, 8), autopct='%1.1f%%', startangle=90)
plt.title('Top 5 developer types')
plt.axis('equal')

plt.legend(labels=top5Typesdf.index, bbox_to_anchor=(1.35, 1), loc='upper right')

plt.show()
```

##### 1.3 Create a pie chart for the operating systems used by respondents for professional use


The `OpSysProfessional` use column shows the operating systems developers use professionally. Let’s visualize the distribution of the top operating systems in a pie chart.



```python
##Write your code here
operatingdf = df['OpSysProfessional use'].reset_index(name='System')
operatingdf.dropna(axis=0, inplace=True)

operatingdf['System'] = operatingdf['System'].str.split(';')

operatingdf = operatingdf.explode('System')

operatingdf = operatingdf.groupby('System').count().sort_values(by='index', ascending=False).head(5)

operatingdf = operatingdf.rename(columns={'index' : 'System'})
operatingdf['System'].plot(kind='pie', figsize=(12, 8), autopct='%1.1f%%', startangle=90)
plt.title('Operating systems used for professional work')
plt.axis('equal')

plt.legend(labels=operatingdf.index, bbox_to_anchor=(1.5, 1), loc='upper right')

plt.show()
```

### Task 2: Additional Visualizations and Comparisons


##### 2.1 Pie Chart for Top 5 Programming Languages Respondents Have Worked With


The `LanguageHaveWorkedWith` column contains the programming languages that respondents have experience with. We’ll plot a pie chart to display the composition of the top 5 languages.



```python
##Write your code here
languagedf = df['LanguageHaveWorkedWith'].reset_index(name='Language')
languagedf.dropna(axis=0, inplace=True)

languagedf['Language'] = languagedf['Language'].str.split(';')

languagedf = languagedf.explode('Language')

languagedf = languagedf.groupby('Language').count().sort_values(by='index', ascending=False).head(5)

languagedf = languagedf.rename(columns={'index' : 'Language'})
languagedf['Language'].plot(kind='pie', figsize=(12, 8), autopct='%1.1f%%', pctdistance=0.9, startangle=90)
plt.title('Languages respondants have worked with')
plt.axis('equal')

plt.legend(labels=languagedf.index, bbox_to_anchor=(1.35, 1), loc='upper right')

plt.show()
```

##### 2.2 Pie Chart for Top Collaboration Tools used in Professional Use


Using the `NEWCollabToolsHaveWorkedWith` column, we’ll identify and visualize the top collaboration tools respondents use in their professional work.



```python
##Write your code here
tooldf = df['NEWCollabToolsHaveWorkedWith'].reset_index(name='Tool')
tooldf.dropna(axis=0, inplace=True)

tooldf['Tool'] = tooldf['Tool'].str.split(';')

tooldf = tooldf.explode('Tool')

tooldf = tooldf.groupby('Tool').count().sort_values(by='index', ascending=False).head(5)

tooldf = tooldf.rename(columns={'index' : 'Tool'})
tooldf['Tool'].plot(kind='pie', figsize=(12, 8), autopct='%1.1f%%', startangle=90)
plt.title('Collaboration tools have worked with')
plt.axis('equal')

plt.legend(labels=tooldf.index, bbox_to_anchor=(1.5, 1), loc='upper right')
plt.show()
```

### Task 3: Analyzing and Interpreting Composition


In this task, you will create additional pie charts to analyze specific aspects of the survey data. Use `pandas` and `matplotlib` to complete each task and interpret the findings.



##### 3.1 Pie Chart of `Respondents` Most Admired Programming Languages


The `LanguageAdmired` column lists the programming languages respondents admire most. Create a pie chart to visualize the top 5 admired languages.



```python
##Write your code here
admiredlanguagedf = df['LanguageAdmired'].reset_index(name='Language')
admiredlanguagedf.dropna(axis=0, inplace=True)

admiredlanguagedf['Language'] = admiredlanguagedf['Language'].str.split(';')

admiredlanguagedf = admiredlanguagedf.explode('Language')

admiredlanguagedf = admiredlanguagedf.groupby('Language').count().sort_values(by='index', ascending=False).head(5)

admiredlanguagedf = admiredlanguagedf.rename(columns={'index' : 'Language'})
admiredlanguagedf['Language'].plot(kind='pie', figsize=(12, 8), autopct='%1.1f%%', startangle=90)
plt.title('Admired Languages')
plt.axis('equal')

plt.legend(labels=admiredlanguagedf.index, bbox_to_anchor=(1.5, 1), loc='upper right')
plt.show()
```

##### 3.2 Pie Chart of Tools Used for AI Development


Using the `AIToolCurrently` Using column, create a pie chart to visualize the top 5 tools developers are currently using for AI development.



```python
##Write your code here
aidf = df['AIToolCurrently Using'].reset_index(name='Ai')
aidf.dropna(axis=0, inplace=True)

aidf['Ai'] = aidf['Ai'].str.split(';')

aidf = aidf.explode('Ai')

aidf = aidf.groupby('Ai').count().sort_values(by='index', ascending=False).head(5)

aidf = aidf.rename(columns={'index' : 'Ai'})
aidf['Ai'].plot(kind='pie', figsize=(12, 8), autopct='%1.1f%%', startangle=90)
plt.title('Tools used for AI development')
plt.axis('equal')

plt.legend(labels=aidf.index, bbox_to_anchor=(1.6, 0.95), loc='upper right')
plt.show()
```

##### 3.3 Pie Chart for Preferred Web Frameworks


The `WebframeWantToWorkWith` column includes web frameworks that respondents are interested in working with. Visualize the top 5 frameworks in a pie chart.



```python
##Write your code here
frameworkdf = df['WebframeWantToWorkWith'].reset_index(name='Framework')
frameworkdf.dropna(axis=0, inplace=True)

frameworkdf['Framework'] = frameworkdf['Framework'].str.split(';')

frameworkdf = frameworkdf.explode('Framework')

frameworkdf = frameworkdf.groupby('Framework').count().sort_values(by='index', ascending=False).head(5)

frameworkdf = frameworkdf.rename(columns={'index' : 'Framework'})
frameworkdf['Framework'].plot(kind='pie', figsize=(12, 8), autopct='%1.1f%%', startangle=90)
plt.title('Frameworks used')
plt.axis('equal')

plt.legend(labels=frameworkdf.index, loc='upper right')
plt.show()
```

##### 3.4 Pie Chart for Most Desired Embedded Technologies


Using the `EmbeddedWantToWorkWith` column, create a pie chart to show the top 5 most desired embedded technologies that respondents wish to work with.



```python
##Write your code here
embaddeddf = df['EmbeddedWantToWorkWith'].reset_index(name='Embedded')
embaddeddf.dropna(axis=0, inplace=True)

embaddeddf['Embedded'] = embaddeddf['Embedded'].str.split(';')

embaddeddf = embaddeddf.explode('Embedded')


embaddeddf = embaddeddf.groupby('Embedded').count().sort_values(by='index', ascending=False).head(5)

embaddeddf = embaddeddf.rename(columns={'index' : 'Embedded'})
embaddeddf['Embedded'].plot(kind='pie', figsize=(12, 8), autopct='%1.1f%%', startangle=90)
plt.title('Embedded technologies respondands want to work with')
plt.axis('equal')

plt.legend(labels=embaddeddf.index, bbox_to_anchor=(1.1, 1), loc='upper right')
plt.show()
```
